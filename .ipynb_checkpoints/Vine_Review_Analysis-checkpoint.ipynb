# Deliverable 2:  Determine Bias of Vine Reviews

### Dependencies and data import

```python
# Dependencies
import pandas as pd

# File to Load
vine_data_to_load = "vine_table.csv"

# Read the Vine Reviews and store into a Pandas DataFrame
vine_data_df = pd.read_csv(vine_data_to_load)

# Check Vine Reviews
vine_data_df.head()
```

```python
vine_data_df.count()
```

### Step 1. Filter the data and create a new DataFrame to retrieve all the rows where the total_votes count is equal to or greater than 20.

```python
filtered_data_df = vine_data_df[(vine_data_df['total_votes'] >= 20)]
filtered_data_df.head()
```

```python
filtered_data_df.count()
```

### Step 2.  Filter the new DataFrame created in Step 1 and create a new DataFrame to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

```python
total_votes_df = filtered_data_df[(filtered_data_df['helpful_votes'] / filtered_data_df['total_votes'] >= 0.5)]
total_votes_df.head()
```

```python
total_votes_df.count()
```

### Step 3.  Filter the DataFrame and create a new DataFrame that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

```python
vine_program_yes_df = total_votes_df[(total_votes_df['vine'] == "Y")]
vine_program_yes_df.head()
```

```python
vine_program_yes_df.count()
```

### Step 4.  Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.

```python
vine_program_no_df = total_votes_df[(total_votes_df['vine'] == "N")]
vine_program_no_df.head()
```

```python
vine_program_no_df.count()
```

### Step 5. Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

* __Determine the total number of reviews__

```python
vine_data_df.count()
```

* __Determine the number of 5-star reviews (Entire dataset)__

```python
five_star_reviews_df = vine_data_df[(vine_data_df['star_rating'] == 5)]
five_star_reviews_df.count()
```

* __Determine the number of 5-star reviews ("Vine" = "Y")__

```python
five_star_reviews_vine_yes_df = five_star_reviews_df[(five_star_reviews_df['star_rating'] == 5) & (five_star_reviews_df['vine'] == "Y")]
five_star_reviews_vine_yes_df.count()
```

* __Determine the number of 5-Star reviews ("Vine" = "N")__

```python
five_star_reviews_vine_no_df = five_star_reviews_df[(five_star_reviews_df['star_rating'] == 5) & (five_star_reviews_df['vine'] == "N")]
five_star_reviews_vine_no_df.count()
```

### Calculations and formatting of totals and percentages.

```python
# Calculations for total reviews, total 5-star reviews, and percentages of Vine and non-Vine 5-star reviews
total_review_count = vine_data_df["star_rating"].count()
five_star_reviews_total = five_star_reviews_df["star_rating"].count()
five_star_vine_reviews = five_star_reviews_vine_yes_df["star_rating"].count()
five_star_reviews = five_star_reviews_vine_no_df["star_rating"].count()
percentage_of_five_star_vine_reviews = five_star_vine_reviews / five_star_reviews_total * 100
percentage_of_five_star_reviews = five_star_reviews / five_star_reviews_total * 100
```

```python
# Create a dataframe to hold the reviews summary
reviews_summary_df = pd.DataFrame(
        [{"Total Reviews": total_review_count,
         "5-Star Reviews": five_star_reviews_total,
         "Vine 5-Star Total": five_star_vine_reviews,
         "non-Vine 5-Star Total": five_star_reviews,
         "Vine 5-Star %age": percentage_of_five_star_vine_reviews,
         "non-Vine 5-Star %age": percentage_of_five_star_reviews}])

# Format the "Total Reviews" to have a comma for a thousands separator.
reviews_summary_df["Total Reviews"] = reviews_summary_df["Total Reviews"].map("{:,}".format)
# Format the "Total 5-Star Reviews" to have a comma for a thousands separator.
reviews_summary_df["5-Star Reviews"] = reviews_summary_df["5-Star Reviews"].map("{:,}".format)
# Format the "Vine 5-Star Total" to have a comma for a thousands separator.
reviews_summary_df["Vine 5-Star Total"] = reviews_summary_df["Vine 5-Star Total"].map("{:,}".format)
# Format the "non-Vine 5-Star Total" to have a comma for a thousands separator.
reviews_summary_df["non-Vine 5-Star Total"] = reviews_summary_df["non-Vine 5-Star Total"].map("{:,}".format)
# Format the "Vine 5-Star %age" to round to the nearest hundredth place.
reviews_summary_df["Vine 5-Star %age"] = reviews_summary_df["Vine 5-Star %age"].map("{:.2f}".format)
# Format the "non-Vine 5-Star %age" to round to the nearest hundredth place.
reviews_summary_df["non-Vine 5-Star %age"] = reviews_summary_df["non-Vine 5-Star %age"].map("{:.2f}".format)

# Drop the data frame index
reviews_summary_df = reviews_summary_df.style.hide_index()

# Display the reviews summary data frame
reviews_summary_df
```
